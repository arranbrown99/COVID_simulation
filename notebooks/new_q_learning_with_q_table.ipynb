## Training and dumping q_tables for all problem ids

```python
for i in range(0,10):
    filename = "q_learning_tabular_problem_id_" + str(i) + ".txt"    
    env = virl.Epidemic(stochastic=False, noisy=False)
    agent = QLearningAgent(env, print_out_every_x_episodes=250)
    agent.train()
    q_table = agent.q_table
    if q_table.save_raw_q_table_to_file(filename):
        print("Saved to file " + filename + " successfully!")
```
